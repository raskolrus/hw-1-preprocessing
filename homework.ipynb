```python
amb= [json.loads(line) for line in open('text_parsed_no.txt', encoding='utf-8')]
```

```python
disamb = [json.loads(line) for line in open('text_parsed_yes.txt', encoding='utf-8')]
```

```python
tok_list = []
amb_list = []
disamb_list = []
for q in amb:
    try:
        amb_list.append(q['analysis'][0]['lex'])
    except:
        amb_list.append('no')
    tok_list.append(q['text'])
for q in disamb:
    try:
        disamb_list.append(q['analysis'][0]['lex'])
    except:
        disamb_list.append('no')
tog = pd.DataFrame()
tog['amb']=amb_list
tog['disamb']=disamb_list
tog['tok']=tok_list
```

```python
tog[tog['amb']!=tog['disamb']]
```
